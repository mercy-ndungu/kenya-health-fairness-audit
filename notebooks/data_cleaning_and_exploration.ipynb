The code that checks the unique values of the status of health units had "non-functinoal" which is a typo
THis indicates that there is data that is wrongly labelled. 
Also, non-functinoal and closed are the same. So the rows that have "non-functinoal" as a status will be turned to Closed.

```python
health_units["Status"] = health_units['Status'].str.replace('non-functinoal', 'Closed')
health_units['Status'].value_counts()
```

```python
#Inspect the dates of Date_established that look sketchy 
print(health_units['Date_established'].head(20).tolist())
```

```python
#See the max and min value of Date_established 
print('\nMin value:', health_units['Date_established'].min())
print('Max value:', health_units['Date_established'].max())
```

The Date_established column has data in number of days as from 1st January 1900 contrary to date time data that is required
Convert it to year_established and drop date, Year_established ans Date_established. 

```python
health_units['date'] = pd.to_datetime(health_units['Date_established'], origin='1899-12-30', unit='D')
health_units['year'] = health_units['date'].dt.year
health_units.drop(['Date_established', 'Year_established', 'date'], axis=1, inplace = True)
health_units.head()
```
