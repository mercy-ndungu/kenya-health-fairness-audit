# Cleaning data

 Clean the status of health units data labeling and standadise it

```python
# Cleaning health_units and create a copy to keep the original dataset
 
hu = health_units.copy()
 
print('Starting shape:', hu.shape)
```

```python
before = hu['Status'].value_counts().to_dict()
 
hu['Status'] = hu['Status'].str.strip().replace({
    'non-functinoal':   'non-functional',
    'Fully-functional': 'fully-functional',
    'semi-functional':  'semi-functional',
    'Closed':           'closed',
    'non-functional':   'closed' 
})
 
after = hu['Status'].value_counts().to_dict()
 
print('Before fix:', before)
print('After fix: ', after)
```

 Convert Date_established from Excel serial to real date

```python

s = pd.to_numeric(hu['Date_established'], errors='coerce')

hu['date'] = pd.to_datetime(
    s.where(s > 3000),  
    origin='1899-12-30',
    unit='D',
    errors='coerce'
)


hu.loc[s <= 3000, 'date'] = pd.to_datetime(
    s[s <= 3000].astype(int),
    format='%Y',
    errors='coerce'
)

hu['year'] = hu['date'].dt.year
hu.drop(columns=['Date_established', 'date'], inplace= True)
hu.head()
```

 Standardise all text columns to lowercase and strip whitespace

```python
# list all text columns
text_columns = ['Facility_ward', 'Facility_constituency', 'Facility_subcounty', 'Facility_county']

for col in text_columns:
    hu[col] = hu[col].str.strip().str.lower()
 
print('Text columns standardised. Sample county values:' ,hu['Facility_county'].unique()[:10])
```

Check for duplicate facilities

```python
#Use health units code as a unique identifier to check for duplicates
duplicate_codes = hu[hu.duplicated('Code', keep=False)]
print(f'Rows with duplicate Code values: {len(duplicate_codes)}')
 
if len(duplicate_codes) > 0:
    print('Dropping duplicates, keeping first occurrence.')
    hu = hu.drop_duplicates('Code', keep='first')
    print('New shape:', hu.shape)
else:
    print('No duplicates found. No rows removed.')
```

```python
#Final check on cleaned health_units
print('Final shape:', hu.shape)
print('\nMissing values:' ,hu.isnull().sum())
print('\nStatus value counts:' ,hu['Status'].value_counts())
```

# Cleaning hospitals_and_beds

```python
#Create a duplicate 
hb = facilities_with_beds.copy()
print('Starting shape:', hb.shape)
```

Convert Beds column from string to integer

```python
before_nulls = hb['Beds'].isnull().sum()
hb['Beds'] = pd.to_numeric(hb['Beds'], errors='coerce')
after_nulls = hb['Beds'].isnull().sum()
 
print(f'Beds column type is now: {hb["Beds"].dtype}')
print(f'Nulls before conversion: {before_nulls}')
print(f'Nulls after conversion:  {after_nulls}')
 
if after_nulls > before_nulls:
    new_nulls = after_nulls - before_nulls
    print(f'{new_nulls} rows had values that could not be converted to numbers.')
```

Investigate zero-bed facilities

```python
zero_beds = hb[hb['Beds'] == 0]
 
print('Facility types with 0 beds:')
print(zero_beds['Facility type'].value_counts().head(15))
 
print(f'\nTotal zero-bed facilities: {len(zero_beds)}')
print(f'Percentage of dataset: {round(len(zero_beds)/len(hb)*100, 1)}%')
```

```python
#Standardise Facility type names 
hb['Facility type'] = hb['Facility type'].str.strip().str.title()
 
#Define which facility types should have beds
inpatient_facility_types = [
    'Primary Care Hospitals',
    'Nursing And Maternity Home',
    'Nursing Homes',
    'Health Centre',
    'Basic Health Centre',
    'Comprehensive Health Centre'
]
 
#Flag only zero-bed rows where the facility type is inpatient
hb['Beds_data_suspect'] = (
    (hb['Beds'] == 0) &
    (hb['Facility type'].isin(inpatient_facility_types))
)
suspect_count = hb['Beds_data_suspect'].sum()
print(f'Inpatient facilities reporting zero beds (suspect): {suspect_count}')
print(f'These are flagged as potentially missing data, not confirmed zeros.')
print()
print('Breakdown by facility type:')
print(
    hb[hb['Beds_data_suspect']]
    ['Facility type']
    .value_counts()
)
```

Drop Service_names

```python
hb = hb.drop(columns=['Service_names'])
print('Dropped Service_names. New shape:', hb.shape)
```

Standardise Yes/No and text columns

```python
yes_no_cols = ['Open_whole_day', 'Open_public_holidays', 'Open_weekends',
               'Open_late_night', 'Approved', 'Public visible', 'Closed']
 
for col in yes_no_cols:
    hb[col] = hb[col].map({'Yes': True, 'No': False})
 
hb['County'] = hb['County'].str.strip().str.lower()
 
print('Standardisation complete.')
hb.head()
```
